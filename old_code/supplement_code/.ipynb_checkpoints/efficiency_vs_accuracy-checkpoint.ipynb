```python
import json
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np

plt.style.use("ggplot")

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
del colors[3]
colors.append(colors[-1])
colors[-2] = "#E6194B"

def plot(ax, method, data, budget, method_idx):
    for i, trial in enumerate(data):
        if trial["per_inst_time_avg"] > budget:
            i = i - 1
            break
    
    trial = data[i]
    if trial["per_inst_time_avg"] > budget:
        ax.plot([], [], c = colors[method_idx + 1], alpha = 0.7, label = label_maps[method])
        return
    
    print(method, trial["config"], trial["per_inst_time_avg"])
    tuples = []
    approximation_data = trial["approximation_data"]
    for inst_idx in approximation_data:
        for layer_idx in approximation_data[inst_idx]:
            for head_idx in approximation_data[inst_idx][layer_idx]:
                for s in approximation_data[inst_idx][layer_idx][head_idx]:
                    d = approximation_data[inst_idx][layer_idx][head_idx][s]
                    tuples.append([d["entropy"], d["error_norm"] / d["output_norm"]])
    tuples = np.asarray(tuples)
    
    pts = np.linspace(0, 8.4, 100)
    entropy = tuples[:, 0]
    err = tuples[:, 1]
    diff = (pts[:, None] - entropy[None, :]) ** 2
    weights = np.exp(- 5 * diff)
    hat_f = np.sum(weights * err[None, :], axis = -1) / np.sum(weights, axis = -1)
    
    ax.plot(pts, hat_f, c = colors[method_idx + 1], alpha = 0.8, linewidth = 2)
    ax.scatter([], [], c = colors[method_idx + 1], label = label_maps[method], alpha = 0.8)
    
```

```python
label_maps = {
    "transformer":"Transformer",
    "nystromformer":"Nystromformer",
    "bigbird":"Big Bird",
    "longformer":"Longformer",
    "scatterbrain":"Scatterbrain",
    "mra-2":"MRA-2",
    "mra-2-s":"MRA-2-s",
}

fig, axes = plt.subplots(1, 2, figsize = (8, 3), sharey = "row")

file = f"outputs/transformer-4096.json"
with open(file, "r") as f:
    data = json.load(f)
assert len(data) == 1
budget_2 = data[0]["per_inst_time_avg"]
budget_4 = data[0]["per_inst_time_avg"] / 2
print(budget_2, budget_4)

methods = ["longformer", "bigbird", "nystromformer", "scatterbrain", "mra-2", "mra-2-s", ]
for method_idx, method in enumerate(methods):
    
    file = f"outputs/{method}-4096.json"
    with open(file, "r") as f:
        data = json.load(f)
    plot(axes[0], method, data, budget_2, method_idx)

    file = f"outputs/{method}-4096.json"
    with open(file, "r") as f:
        data = json.load(f)
    plot(axes[1], method, data, budget_4, method_idx)

axes[0].set_xlabel("Entropy", fontsize = 14)
axes[0].set_ylabel("Relative Error", fontsize = 14)
axes[0].set_ylim([0, 1.5])
axes[0].set_title("< 30 ms", fontsize = 14, alpha = 0.7)

axes[1].set_xlabel("Entropy", fontsize = 14)
axes[1].set_title("< 15 ms", fontsize = 14, alpha = 0.7)

axes[0].tick_params(labelsize = 14)
axes[1].tick_params(labelsize = 14)

fig.tight_layout()

lgd = axes[0].legend(
    handletextpad = 0.1, columnspacing = 1, fontsize = 12, 
    loc = 'upper center', bbox_to_anchor = (0.95, 1.38), ncol = 6)

fig.savefig("entropy_vs_accuracy.pdf", dpi = 400, bbox_extra_artists=(lgd, ), bbox_inches='tight')
```

```python
import json
import matplotlib.pyplot as plt
import os
import pandas as pd

plt.style.use("ggplot")

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
del colors[3]
colors.append(colors[-1])
colors[-2] = "#E6194B"

def plot_line(ax, x, y, s, label, color):
    ax.scatter(x, y, s = s, color = color, alpha = 0.8)
    ax.plot(x, y, linestyle = '-', color = color, alpha = 0.8, linewidth = 2)
    ax.scatter([], [], color = color, label = label, alpha = 0.8)

def plot(ax, method, df, marker_size, method_idx):
    tuples = []
    for i, trial in enumerate(data):
        time = trial["per_inst_time_avg"]
        mem = trial["memory_per_inst"]
        
        error = []
        approximation_data = trial["approximation_data"]
        for inst_idx in approximation_data:
            for layer_idx in approximation_data[inst_idx]:
                for head_idx in approximation_data[inst_idx][layer_idx]:
                    d = approximation_data[inst_idx][layer_idx][head_idx]["1"]
                    error.append(d["error_norm"] / d["output_norm"])
        error = np.mean(error).item()
        tuples.append([time, mem, error])
    tuples = np.asarray(tuples)
    
    if method == "transformer":
        t = tuples[0, 0].item()
        ax.plot([t] * 2, [0, 2], linestyle='--', color = colors[method_idx], alpha = 0.7, linewidth = 2)
    else:
        plot_line(ax, tuples[:, 0], tuples[:, 2], marker_size * tuples[:, 1], label_maps[method], colors[method_idx])
```
